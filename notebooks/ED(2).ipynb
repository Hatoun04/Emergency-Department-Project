```python
## imports for modeling
#!pip show statsmodels
%pip install shap


import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
import statsmodels.api as sm
from statsmodels.tsa.statespace.sarimax import SARIMAX
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
import shap
```

Data Preparation

```python
# Aggregate visits into hourly time buckets
data['Visit Hour'] = pd.to_datetime(data['Visit Date']).dt.floor('h')

# Aggregate features per Hospital and Visit Hour
agg_data = data.groupby(['Hospital ID', 'Visit Hour']).agg({
    'Visit Date': 'first',
    'Region': 'first', 'Season': 'first', 'Day of Week': 'first', 'Time of Day': 'first',
    'Urgency Level': 'mean', 'Nurse-to-Patient Ratio': 'mean',
    'Specialist Availability': 'mean', 'Facility Size (Beds)': 'first',
    'hour': 'first', 'weekday_num': 'first', 'month': 'first',
    'wait_lag_1w': 'mean', 'wait_rolling_3d': 'mean', 'urgency_x_nurse': 'mean',
    'Total Wait Time (min)': 'mean'
}).reset_index()
```

```python
# Target variable: number of visits per hospital-hour
agg_data['Visit Count'] = data.groupby(['Hospital ID', 'Visit Hour']).size().values

# Set index and fill missing values
agg_data.set_index('Visit Date', inplace=True)
agg_data = agg_data.sort_index().asfreq('h', method='ffill')
agg_data.fillna({'wait_lag_1w': 0, 'wait_rolling_3d': 0, 'urgency_x_nurse': 0}, inplace=True)
```

```python
# Chronological split (70% train, 15% validation, 15% test)
n = len(agg_data)
train_end = int(0.7 * n)
val_end = int(0.85 * n)
train_agg = agg_data.iloc[:train_end]
val_agg = agg_data.iloc[train_end:val_end]
test_agg = agg_data.iloc[val_end:]
print(f"Shapes: Train {train_agg.shape}, Val {val_agg.shape}, Test {test_agg.shape}")
```

Baseline Model: SARIMA

```python
def fit_sarima(ts):
    """Fit a SARIMA model to a univariate hospital time series."""
    try:
        ts = ts.asfreq('h', method='ffill')
        model = SARIMAX(ts, order=(1,1,1), seasonal_order=(1,1,1,24),
                        enforce_stationarity=False)
        return model.fit(disp=False)
    except:
        return None
```

```python
# Train SARIMA per hospital
sarima_models = {
    hid: fit_sarima(group['Visit Count'])
    for hid, group in train_agg.groupby('Hospital ID')
}
sarima_models = {hid: m for hid, m in sarima_models.items() if m is not None}
```

```python
# Forecast for Validation set
sarima_preds_val = {}
for hid, model in sarima_models.items():
    val_group = val_agg[val_agg['Hospital ID'] == hid]
    if len(val_group) > 0:
        sarima_preds_val[hid] = model.forecast(len(val_group))
```

```python
# Forecast for Test set
sarima_preds_test = {}
for hid, model in sarima_models.items():
    test_group = test_agg[test_agg['Hospital ID'] == hid]
    if len(test_group) > 0:
        sarima_preds_test[hid] = model.forecast(len(test_group))
```

```python
#Evaluate SARIMA
if sarima_preds_val:
    sarima_val_mae = mean_absolute_error(
        val_agg[val_agg['Hospital ID'].isin(sarima_preds_val.keys())]['Visit Count'],
        pd.concat(sarima_preds_val.values())
    )
    sarima_val_mape = mean_absolute_percentage_error(
        val_agg[val_agg['Hospital ID'].isin(sarima_preds_val.keys())]['Visit Count'],
        pd.concat(sarima_preds_val.values())
    )
    sarima_test_mae = mean_absolute_error(
        test_agg[test_agg['Hospital ID'].isin(sarima_preds_test.keys())]['Visit Count'],
        pd.concat(sarima_preds_test.values())
    )
    sarima_test_mape = mean_absolute_percentage_error(
        test_agg[test_agg['Hospital ID'].isin(sarima_preds_test.keys())]['Visit Count'],
        pd.concat(sarima_preds_test.values())
    )
    print(f"SARIMA Val -> MAE: {sarima_val_mae:.2f}, MAPE: {sarima_val_mape:.2%}")
    print(f"SARIMA Test -> MAE: {sarima_test_mae:.2f}, MAPE: {sarima_test_mape:.2%}")
else:
    print("SARIMA: No valid predictions")
```

Advanced Multivariate Model: XGBoost

```python
# Prepare features/target
X_train = train_agg.drop(['Visit Count', 'Hospital ID', 'Visit Hour'], axis=1)
y_train = train_agg['Visit Count']
X_val = val_agg.drop(['Visit Count', 'Hospital ID', 'Visit Hour'], axis=1)
y_val = val_agg['Visit Count']
X_test = test_agg.drop(['Visit Count', 'Hospital ID', 'Visit Hour'], axis=1)
y_test = test_agg['Visit Count']
```

```python
# Fit XGBoost
xgb_model = xgb.XGBRegressor(objective='reg:squarederror',
                             n_estimators=100, learning_rate=0.05, max_depth=5)
xgb_model.fit(X_train, y_train)
```

```python
# Predictions
xgb_preds_val = xgb_model.predict(X_val)
xgb_preds_test = xgb_model.predict(X_test)
```

```python
# Evaluation
xgb_val_mae = mean_absolute_error(y_val, xgb_preds_val)
xgb_val_mape = mean_absolute_percentage_error(y_val, xgb_preds_val)
xgb_test_mae = mean_absolute_error(y_test, xgb_preds_test)
xgb_test_mape = mean_absolute_percentage_error(y_test, xgb_preds_test)
print(f"XGBoost Val -> MAE: {xgb_val_mae:.2f}, MAPE: {xgb_val_mape:.2%}")
print(f"XGBoost Test -> MAE: {xgb_test_mae:.2f}, MAPE: {xgb_test_mape:.2%}")
```

Hybrid Model: SARIMA + XGBoost

```python
# Add SARIMA residuals to training data
train_agg = train_agg.copy()
train_agg['sarima_resid'] = 0.0
for hid, model in sarima_models.items():
    train_group = train_agg[train_agg['Hospital ID'] == hid]
    if model is not None:
        train_agg.loc[train_group.index, 'sarima_resid'] = (
            y_train.loc[train_group.index] - model.fittedvalues
        )
```

```python
# Add SARIMA predictions to validation/test sets
val_agg = val_agg.copy()
val_agg['sarima_pred'] = np.nan
for hid, preds in sarima_preds_val.items():
    val_agg.loc[val_agg['Hospital ID'] == hid, 'sarima_pred'] = preds

test_agg = test_agg.copy()
test_agg['sarima_pred'] = np.nan
for hid, preds in sarima_preds_test.items():
    test_agg.loc[test_agg['Hospital ID'] == hid, 'sarima_pred'] = preds
```

```python
# Build hybrid features (Val)
X_val_hybrid = val_agg.drop(['Visit Count', 'Hospital ID', 'Visit Hour', 'sarima_pred'], axis=1).copy()
X_val_hybrid['sarima_resid'] = y_val - val_agg['sarima_pred'].fillna(0)
```

```python
# Build hybrid features (Test)
X_test_hybrid = test_agg.drop(['Visit Count', 'Hospital ID', 'Visit Hour', 'sarima_pred'], axis=1).copy()
X_test_hybrid['sarima_resid'] = y_test - test_agg['sarima_pred'].fillna(0)
```

```python
# Fit XGBoost on training data (with SARIMA residuals)
hybrid_model = xgb.XGBRegressor(objective='reg:squarederror',
                                n_estimators=100, learning_rate=0.05, max_depth=5)
hybrid_model.fit(train_agg.drop(['Visit Count', 'Hospital ID', 'Visit Hour'], axis=1), y_train)
```

```python
# Hybrid predictions
hybrid_preds_val = hybrid_model.predict(X_val_hybrid) + val_agg['sarima_pred'].fillna(0)
hybrid_preds_test = hybrid_model.predict(X_test_hybrid) + test_agg['sarima_pred'].fillna(0)
```

Evaluation

```python
hybrid_val_mae = mean_absolute_error(y_val, hybrid_preds_val)
hybrid_val_mape = mean_absolute_percentage_error(y_val, hybrid_preds_val)
hybrid_test_mae = mean_absolute_error(y_test, hybrid_preds_test)
hybrid_test_mape = mean_absolute_percentage_error(y_test, hybrid_preds_test)
print(f"Hybrid Val -> MAE: {hybrid_val_mae:.2f}, MAPE: {hybrid_val_mape:.2%}")
print(f"Hybrid Test -> MAE: {hybrid_test_mae:.2f}, MAPE: {hybrid_test_mape:.2%}")
```

Retrain XGBoost on train + val (clean feature set)

```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# 1) Build clean feature matrices (drop SARIMA artifacts if they exist)
drop_always = ['Visit Count','Hospital ID','Visit Hour']
sarima_artifacts = ['sarima_resid','sarima_pred','sarima_fitted']

def make_Xy(df):
    X = df.drop(columns=[c for c in (drop_always + sarima_artifacts) if c in df.columns])
    y = df['Visit Count']
    return X, y

trainval_agg = pd.concat([train_agg, val_agg]).sort_values(['Hospital ID','Visit Hour'])
X_trainval, y_trainval = make_Xy(trainval_agg)
X_test_,     y_test_    = make_Xy(test_agg)

# Make sure test has same columns/order as trainval
X_test_ = X_test_.reindex(columns=X_trainval.columns, fill_value=0)

# 2) Retrain XGBoost
xgb_best = xgb.XGBRegressor(
    objective='reg:squarederror',
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.9,
    colsample_bytree=0.9,
    random_state=42,
)
xgb_best.fit(X_trainval, y_trainval)

# 3) Evaluate on test
yhat_test = xgb_best.predict(X_test_)
print(
    f"[XGB retrained] Test MAE: {mean_absolute_error(y_test_, yhat_test):.4f} | "
    f"MAPE: {mean_absolute_percentage_error(y_test_, yhat_test):.2%}"
)
```

Uncertainty (Conformal 90% interval)

```python
import numpy as np
import pandas as pd

# Use your existing validation predictions if available, else compute quickly:
try:
    y_true_val = y_val.to_numpy()
    y_pred_val = np.asarray(xgb_preds_val)  # from your earlier fit on train
except NameError:
    # fallback: quickly train on train only to get val preds (keeps calibration separate)
    X_train = train_agg.drop(columns=[c for c in ['Visit Count','Hospital ID','Visit Hour'] if c in train_agg.columns])
    y_train = train_agg['Visit Count']
    X_val_  = val_agg.drop(columns=[c for c in ['Visit Count','Hospital ID','Visit Hour'] if c in val_agg.columns])
    y_val_  = val_agg['Visit Count']
    xgb_cal = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=200, learning_rate=0.05, max_depth=6, random_state=42)
    xgb_cal.fit(X_train, y_train)
    y_true_val = y_val_.to_numpy()
    y_pred_val = xgb_cal.predict(X_val_)

alpha = 0.10  # 90% PI
q90 = np.quantile(np.abs(y_true_val - y_pred_val), 1 - alpha)

yhat_lo_90 = yhat_test - q90
yhat_hi_90 = yhat_test + q90

uncertainty_df = test_agg[['Hospital ID','Visit Hour']].copy()
uncertainty_df['yhat'] = yhat_test
uncertainty_df['yhat_lo_90'] = yhat_lo_90
uncertainty_df['yhat_hi_90'] = yhat_hi_90
display(uncertainty_df.head())
```

Staffing recommendations (point & safe)

```python
import numpy as np
import pandas as pd

plan = test_agg[['Hospital ID','Visit Hour','Urgency Level','Specialist Availability']].copy()
plan['yhat']     = yhat_test
plan['yhat_lo']  = yhat_lo_90
plan['yhat_hi']  = yhat_hi_90

# Peak threshold per hospital (top 20% predicted demand)
plan['peak'] = (
    plan.groupby('Hospital ID')['yhat']
        .transform(lambda s: s >= s.quantile(0.80))
).astype(int)

# Policy knobs (tune with ops team)
TARGET_MIN_PER_PATIENT = 20           # minutes of nursing per patient per hour-block
SHIFT_HOURS = 1                       # block size
MIN_NURSE_PATIENT_RATIO_PEAK = 0.5    # 1 nurse per 2 patients during peaks

def nurses_point(r):
    demand = max(0.0, r['yhat'])
    base   = np.ceil(demand * TARGET_MIN_PER_PATIENT / 60 / SHIFT_HOURS)
    urgency = float(r.get('Urgency Level', 0.5))
    avail   = float(r.get('Specialist Availability', 0.5))
    adj = 1.0 + 0.5*urgency - 0.2*avail   #  urgency adds staff,  specialists reduces a bit
    n = max(1, int(np.ceil(base * adj)))
    if r['peak'] == 1 and demand > 0:
        n = max(n, int(np.ceil(demand * MIN_NURSE_PATIENT_RATIO_PEAK)))
    return n

def nurses_safe(r):
    demand_hi = max(0.0, r['yhat_hi'])   # plan to upper confidence bound
    base   = np.ceil(demand_hi * TARGET_MIN_PER_PATIENT / 60 / SHIFT_HOURS)
    urgency = float(r.get('Urgency Level', 0.5))
    avail   = float(r.get('Specialist Availability', 0.5))
    adj = 1.0 + 0.5*urgency - 0.2*avail
    n = max(1, int(np.ceil(base * adj)))
    if r['peak'] == 1 and demand_hi > 0:
        n = max(n, int(np.ceil(demand_hi * MIN_NURSE_PATIENT_RATIO_PEAK)))
    return n

def specialists_needed(r):
    # heuristic: plan specialists to support complex/urgent cases
    demand_ref = r['yhat_hi']  # be conservative
    avail  = float(r.get('Specialist Availability', 0.5))
    base_share = 0.15 if avail >= 0.4 else 0.30
    return int(np.ceil(max(0.0, demand_ref * base_share)))

plan['nurses_recommended']      = plan.apply(nurses_point, axis=1)
plan['nurses_recommended_safe'] = plan.apply(nurses_safe,  axis=1)
plan['specialists_recommended'] = plan.apply(specialists_needed, axis=1)

staffing_plan = plan[['Hospital ID','Visit Hour','yhat','yhat_lo','yhat_hi','peak',
                      'nurses_recommended','nurses_recommended_safe','specialists_recommended']] \
                .rename(columns={'yhat_lo':'yhat_lo_90','yhat_hi':'yhat_hi_90'})

display(staffing_plan.head(10))
```
